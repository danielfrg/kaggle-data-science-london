```python
X = X_df.values
```

```python
y = y_df[0].values
```

### Normalize

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
Xs = scaler.fit_transform(X)
```

```python
Xs_df = pd.DataFrame(Xs)
```

```python
Xs_df.describe().ix['mean'].hist()
```

```python
Xs_df.describe().ix['std'].hist()
```

```python
Xs_df.describe().ix['min'].hist()
```

```python
Xs_df.to_csv('../data/train_inputs_scaled.csv', index=None, header=None)
```
